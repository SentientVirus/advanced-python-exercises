### Step 6. Print the mean alcoohol consumption per continent for every column

```python
drinks.groupby('continent').mean()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit_mean = drinks.groupby('continent').mean()[['spirit_servings']]
spirit_min = drinks.groupby('continent').min()[['spirit_servings']]
spirit_max = drinks.groupby('continent').max()[['spirit_servings']]
print("The mean consumption is:")
print(spirit_mean, "\n\n")
print("The min consumption is:")
print(spirit_min, "\n\n")
print("The max consumption is:")
print(spirit_max)
```
